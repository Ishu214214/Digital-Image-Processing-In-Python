```python
img=cv2.imread('cat.jpg')
plt.imshow(img)
```

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray1 = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
rgb2=cv2.cvtColor(gray1, cv2.COLOR_GRAY2RGB)
rgb3=cv2.cvtColor(gray1, cv2.COLOR_RGB2BGR)

plt.subplot(221)
plt.imshow(rgb)


plt.subplot(222)
plt.imshow(gray1)

plt.subplot(223)
plt.imshow(rgb2)

plt.subplot(224)
plt.imshow(rgb3)
```
